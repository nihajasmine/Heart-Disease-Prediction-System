```python
# Import libraries
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

warnings.filterwarnings("ignore")

# Load the dataset
df = pd.read_csv("heart.csv")
print(df.head())
print("\nMissing values before preprocessing:\n", df.isnull().sum())
```

```python
# Split independent/dependent variables
x = df[[
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
    "thalach", "exang", "oldpeak", "slope", "ca", "thal"
]]
y = df["target"]

# Encode categorical columns
le = LabelEncoder()
categorical_cols = ["sex", "cp", "fbs", "restecg", "exang", "slope", "ca", "thal"]
for col in categorical_cols:
    x[col] = le.fit_transform(x[col].astype(str))

print("\nMissing values after preprocessing:\n", x.isnull().sum())

# Train-test split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
# Initialize models
models = {
    "Decision Tree": DecisionTreeClassifier(random_state=0),
    "KNN": KNeighborsClassifier(n_neighbors=3),
    "SVM": SVC(kernel='linear', random_state=0),
    "Naive Bayes": GaussianNB()
}

# Train & evaluate models
results = []

for name, model in models.items():
    model.fit(train_x, train_y)
    y_pred = model.predict(test_x)

    acc = accuracy_score(test_y, y_pred)
    prec = precision_score(test_y, y_pred)
    rec = recall_score(test_y, y_pred)
    f1 = f1_score(test_y, y_pred)

    results.append({
        "Model": name,
        "Accuracy": round(acc * 100, 2),
        "Precision": round(prec * 100, 2),
        "Recall": round(rec * 100, 2),
        "F1-Score": round(f1 * 100, 2)
    })

# Display performance comparison
performance_df = pd.DataFrame(results)
print("\nModel Performance Comparison:\n")
print(performance_df.to_string(index=False))

# Plot confusion matrices
plt.figure(figsize=(10, 8))
for i, (name, model) in enumerate(models.items(), 1):
    y_pred = model.predict(test_x)
    cm = confusion_matrix(test_y, y_pred)
    plt.subplot(2, 2, i)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f"{name} Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")

plt.tight_layout()
plt.show()
```

```python
# --- Bar chart comparison for model performance ---

plt.figure(figsize=(10, 6))

# Bar chart for Accuracy
plt.subplot(1, 2, 1)
sns.barplot(x="Model", y="Accuracy", data=performance_df)
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy (%)")
plt.xticks(rotation=20)

# Bar chart for F1-Score
plt.subplot(1, 2, 2)
sns.barplot(x="Model", y="F1-Score", data=performance_df)
plt.title("Model F1-Score Comparison")
plt.ylabel("F1-Score (%)")
plt.xticks(rotation=20)

plt.tight_layout()
plt.show()
```

```python
# Save the Decision Tree model only
dt_model = models["Decision Tree"]

import pickle

# Save model
with open("heart.pkl", "wb") as f:
    pickle.dump(dt_model, f)

# Load model back
with open("heart.pkl", "rb") as f:
    loaded_dt_model = pickle.load(f)

# Optional: verify it works
y_pred = loaded_dt_model.predict(test_x)
print("Accuracy after loading:", accuracy_score(test_y, y_pred))
```
